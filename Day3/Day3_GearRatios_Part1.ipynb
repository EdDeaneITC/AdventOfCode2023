```python
from neo4j import GraphDatabase
import re
advent = "Day3"
workdir = 'C:\\Users\\ed.deane\\OneDrive - Sabio Ltd\\Documents\\Customers\\IKEA\\Knowledge\\Python\\AdventOfCode\\2023\\'

#create a node for each Game (id=n)
uri = "bolt://localhost:7687"
username = "neo4j"
password = "advent_of_code"
# Cypher query to create a node with the index as the node name and value as the 'prop' property
query_del = "MATCH (n) DETACH DELETE n"
query_num = "CREATE (:num {name:$name, num:$num, x_start:$x_start,x_end:$x_end, line_num:$line_num})"
query_symbol = "CREATE (:symbol {name:$name, symbol:$symbol, x_start:$x_start,x_end:$x_end, line_num:$line_num})"

with GraphDatabase.driver(uri, auth=(username, password)) as driver:
    with driver.session() as session:
            session.run(query_del)
            
lines = []
f = open(workdir+advent+"\\input_1.txt", "r")
for line in f:
    lines.append (line.strip())
    
def create_num_node(tx, num, x, line_num):
    name = str(line_num)+'_'+str(x)+'_'+num
    tx.run(query_num, name=name, num = num, x_start=x,x_end=x+len(num)-1, line_num=line_num)
    
def create_symbol_node(tx, symbol, x, line_num):
    name = str(line_num)+'_'+str(x)+'_'+symbol
    tx.run(query_symbol, name=name,symbol=symbol, x_start=x,x_end=x+len(symbol)-1, line_num=line_num)
    
def get_numbers(line):
    #return unique list of numbers from the line
    nums = []
    #replace all non numerics with dots
    line = re.sub("[^0-9]", ".",line)
    for num in re.finditer(r'\d+', line):
        nums.append(num[0])
    return list(set(nums))

def get_symbols(line):
    symbols = []
    #replace all non numberics with dots
    line = re.sub("[0-9]", ".",line)
    
    for sym in line.split('.'):
        if len(sym) > 0:
            symbols.append(sym)
    return list(set(symbols))
#identfy each unique number from the line and add to a node that also include a 1st digit location (x,line)

with GraphDatabase.driver(uri, auth=(username, password)) as driver:
    with driver.session() as session:
        for line_num, l in enumerate(lines):
            print (l)
            for num in (get_numbers(l)):
                
                for num_pos in [m.start() for m in re.finditer(num, l)]:
                    print (num,num_pos,line_num)
                    session.write_transaction(create_num_node,num,num_pos,line_num)
                    
            
            for symbol in (get_symbols(l)):
                print ('symbol',symbol)
                for symbol_pos in [m.start() for m in re.finditer(re.escape(symbol), l)]:
                    print (symbol,symbol_pos,line_num)
                    session.write_transaction(create_symbol_node,symbol,symbol_pos,line_num)

#identify each symbol and add a node that also includes a location (x,line)

#link each number to a symbol where the perimeter overlaps
```
